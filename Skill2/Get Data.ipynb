```python
import requests
import pandas as pd

# Make a GET request to a specific endpoint
url = 'https://api.tfl.gov.uk/AccidentStats/2018'

response = requests.get(url)

data = response.json()

data[0]

for entry in data:
  id = entry['id']
  location = entry['location']
  time = entry['date']
  lat = entry['lat']
  lon = entry['lon']

df = pd.DataFrame(data, columns=['id','location','date','lat','lon'])
df
```
